```python
import matplotlib.pyplot as plt
import numpy as np
# see https://pyprog.pro/mpl/mpl_main_components.html
# and https://www.w3resource.com/graphics/matplotlib/
```

```python
x = np.linspace(0,10,10)
y = x**2
```

```python
fig = plt.figure()
lines = plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)

plt.show()
```

Вообще, было бы правильнее использовать fig.add_axes

но в данном случае fig.add_subplot(111) удобнее, т.к. subplot просто размещает Axes на сетке Figure. 

add_subplot(111) - это:
1я строка, 1й столбец, 1я (и единственная) ячейка на сетке Figure.

В том что Figure и Axes это разные области можно легко убедиться если изменить их цвет:

```python
fig = plt.figure()
ax = fig.add_subplot(111)

fig.set(facecolor = 'green')
ax.set(facecolor = 'red')
ax.plot(x,y)

plt.show()
```

```python
# можно сделать это иначе
fig = plt.figure()
ax = fig.add_subplot(111)

fig.set_facecolor('green')
ax.set(facecolor = 'red')
ax.plot(x,y)

plt.show()
```

```python
# матлаб позволяет проникать клубоко во внутренности каждой сущности - например, Axes

fig = plt.figure()
ax = fig.add_subplot(111) 

fig.set_facecolor('green')

ax.set_facecolor('red')
ax.set_xlim([-10, 10])
ax.set_ylim([-100, 100])
ax.set_title('Основы анатомии matplotlib')
ax.set_xlabel('ось абцис (XAxis)')
ax.set_ylabel('ось ординат (YAxis)')
ax.plot(x,y)

plt.show()
```

```python
# но проще, конечно, делать это с помощью более верхнеуровневых методов
# например, метод set() есть почти у всех объектов matplotlib

fig = plt.figure()
ax = fig.add_subplot(111)

fig.set(facecolor = 'green')
ax.set(facecolor = 'red',
       xlim = [-10, 10],
       ylim = [-100, 100],
       title = 'Основы анатомии matplotlib',
       xlabel = 'ось абцис (XAxis)',
       ylabel = 'ось ординат (YAxis)')

ax.plot(x,y)

plt.show()
```

```python
# параметры внешнего вида графиков обычно указываются в аргументах функции plot или scatter

# и, кстати, магической командной можно сделать графики интерактивными
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(111)

fig.set(facecolor = 'green')
ax.set(facecolor = 'red',
       xlim = [-10, 10],
       ylim = [-100, 100],
       title = 'Основы анатомии matplotlib',
       xlabel = 'ось абцис (XAxis)',
       ylabel = 'ось ординат (YAxis)')

ax.plot(x,y, '.k', markersize=20)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

fig.set(facecolor = 'green')
ax.set(facecolor = 'red',
       title = 'Основы анатомии matplotlib',
       xlabel = 'ось абцис (XAxis)',
       ylabel = 'ось ординат (YAxis)')

# каждой точке можно задать цвет
# цветовые палитры matplotlib есть по этой ссылке
# https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

scatter_plot = ax.scatter(x,y, c=y, cmap="viridis")
# саму ось можно добавить, а потом ещё и как угодно расположить на графике
fig.colorbar(scatter_plot)

plt.show()
```

```python
# можно размер точек сделать зависимым от значения
fig = plt.figure()
ax = fig.add_subplot(111)

fig.set(facecolor = 'green')
ax.set(facecolor = 'red',
       title = 'Основы анатомии matplotlib',
       xlabel = 'ось абцис (XAxis)',
       ylabel = 'ось ординат (YAxis)')

# каждой точке можно задать цвет
# цветовые палитры matplotlib есть по этой ссылке
# https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

scatter_plot = ax.scatter(x,y, s=y, c='k')

plt.show()
```

```python
# можно вывести несколько графиков

# а другой магической командой - обратно обычными картинками
%matplotlib inline

fig = plt.figure()

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)
ax_4 = fig.add_subplot(2, 2, 4)

ax_1.set(title = 'ax_1', xticks=[], yticks=[])
ax_2.set(title = 'ax_2', xticks=[], yticks=[])
ax_3.set(title = 'ax_3', xticks=[], yticks=[])
ax_4.set(title = 'ax_4', xticks=[], yticks=[])

plt.show()
```

```python
# причём с размещением в желаемых зонах и желаемых пропорциях

fig = plt.figure()

ax_1 = fig.add_subplot(3, 1, 1)
ax_2 = fig.add_subplot(3, 2, 4)
ax_3 = fig.add_subplot(3, 3, 9)

ax_1.set(title = 'ax_1', xticks=[], yticks=[])
ax_2.set(title = 'ax_2', xticks=[], yticks=[])
ax_3.set(title = 'ax_3', xticks=[], yticks=[])

plt.show()
```
