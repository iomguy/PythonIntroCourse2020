# Проверка строки на символы
Проверить, что в строке содержатся символы только из некоторого множества (например, буквы a-z, A-Z и цифры 0-9).
<!--  https://www.w3resource.com/python-exercises/re/python-re-exercise-1.php -->

## Что нужно сделать?



```python
import re
def is_allowed_specific_char(string):
#     ^ обозначает, что подойдут ЛЮБЫЕ символы кроме тех, что будут дальше
    charRe = re.compile(r'[^a-zA-Z0-9]')
    string = charRe.search(string)
    return not bool(string), string
```

```python
is_allowed_specific_char("ABCDEFabcdef123450")
```

```python
is_allowed_specific_char("*&%@#!}{")
```

```python
is_allowed_specific_char("ADFB%")
```

# Найти все наречия в строке
Наречия в английском заканчиваются на -ly

```python
import re

text = "Clearly, he has no excuse for such behavior. Absolutely"

for m in re.finditer(r"\w+ly", text):
    print(m.start(), m.end(), m.group(0))
    
# метод group есть у многих объектов, которыми оперирует re
# в нём лежат все "подошедние" под шаблон подстроки
```
